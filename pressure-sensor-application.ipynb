## Общие сведения

В документе представлено техническое задание на разработку системы измерения давления жидкости в трубке и ее использование в масштабе распределенной системы.

### Назначение и цели создания системы

1. Определение производительности насоса и расхода жидкости.
2. Сбор данных и последующая их статистическая обработка на серверной стороне для корректировки модели управления и предсказания проблем с оборудованием на основе статистического анализа.

### Требования к системе

1. Прозрачность используемых алгоритмов обработки сигнала с датчика.
2. Отделение логики целевой системы (бизнес-логики) от аппаратной части для реализации приложения на разных процессорных архитектурах микроконтроллеров (ARM Cortex, AVR, RISC-V).


### Блок-схема работы реального приложения

![Блок-схема работы приложения](figures/tz/bs1.png)

На рисунке представлена блок-схема общей работы реального приложения в контексте распределенной системы микроконтроллеров, соединенной между собой через шину CAN. Микроконтроллер забирает данные с датчика, производит быструю обработку данных (фильтрацию), совершает нужные действия с отфильтрованными данными (например, отрисовывает значение давления на дисплее) и, в общем случае, передает данные в шину CAN, где с ней будут работать другие микроконтроллеры. Из шины CAN, в том числе, получает данные микроконтроллер с WI-FI/Bluetooth и отсылает данные на телефон пользователя, откуда данные уходят на сервер для постобработки.

Таким образом, система должна быть модульной и расширяемой.

### Прототип системы измерения расхода жидкости.

В качестве одного из модулей этой системы можно рассмотреть устройство для измерения расхода топлива в трубке по показаниям датчика давления.
Используется датчик давления LPS22HH, соединенный с микроконтроллером STM32L100RCT6 через шину SPI.
